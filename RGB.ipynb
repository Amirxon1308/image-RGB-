```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
IMG_PATH =r"C:\Users\Student\Desktop\1_tAGoymnLMRkEjA2fElG3yQ.jpeg"

imgArray = cv.imread(IMG_PATH)
```

```python
plt.imshow(imgArray)
plt.show()
```

```python
convertedArray = cv.cvtColor(imgArray, cv.COLOR_BGR2RGB)

plt.subplots(figsize=(15,10))
plt.imshow(convertedArray);plt.show()
```

```python
fig,((ax1, ax2),(ax3,ax4)) = plt.subplots(2,2,figsize=(14,10))

ax1.imshow(convertedArray[:,:,0], cmap="Reds_r"); ax1.set_title("R",size=20)
ax2.imshow(convertedArray[:,:,1], cmap="Greens_r"); ax2.set_title("G",size=20)
ax3.imshow(convertedArray[:,:,2], cmap="Blues_r"); ax3.set_title("B",size=20)

ax4.axis("off"); plt.tight_layout(); plt.show()
```

```python
fig,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(15,4))

ax1.hist(convertedArray[:,:,0].flatten(), color="Red",bins=200); ax1.set_title("R",size=20)
ax2.hist(convertedArray[:,:,1].flatten(), color="Green",bins=200); ax2.set_title("G",size=20)
ax3.hist(convertedArray[:,:,2].flatten(), color="Blue",bins=200); ax3.set_title("B",size=20)

plt.tight_layout();plt.show()
```
